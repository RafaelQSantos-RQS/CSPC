## Criando nmero aleatrios

```python
import pandas as pd
import random
```

```python
def criando_dataframe():
    df_dict = {'Caso': [], 'Pessoas': [], 'Salto': []}

    for i in range(30):

        n = random.randint(1,10000) # Gera o numero de pessoas nmero nmero aleatrio de pessoas
        k = random.randint(1,min(1000,n-1)) # O numero de salto no pode ser maior que o nmero de pessoas

        df_dict['Caso'].append(i+1)
        df_dict['Pessoas'].append(n)
        df_dict['Salto'].append(k)

        df = pd.DataFrame(df_dict)
    return df

df = criando_dataframe()
```

```python
def Flavius_Josephus(n,k):
    # Criao de lista para imaginar a circulo de homens
    lista = [x for x in range(1,n+1)]

    posicao = 0
    while len(lista) > 1:
        # Atualizando a varivel posio
        posicao = (posicao + k-1) % len(lista)

        # Eliminando o valor que  o "Homem"
        lista.pop(posicao)

    return lista[0]
```

```python
df['ltimo'] = df.apply(lambda row: Flavius_Josephus(row['Pessoas'], row['Salto']), axis=1)
```

```python
for index,row in df.iterrows():
    print(f"No caso {row['Caso']} so {row['Pessoas']} pessoas, o tamanho do salto  de {row['Salto']} pessoas e o ltimo homem a sobreviver  {row['ltimo']}")
```
